```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/USD_CHF Historical Data.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data = data.drop(columns=['Vol.'])
```

```python
data.shape
```

```python

data['Change %'] = data['Change %'].str.replace('%', '').astype(float) / 100
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['MA_10'] = data['Price'].rolling(window=10).mean()
data['MA_50'] = data['Price'].rolling(window=50).mean()
```

```python

data['Target'] = data['Price'].shift(-1)


data.dropna(inplace=True)
```

```python
X = data[['Price', 'Open', 'High', 'Low', 'MA_10', 'MA_50', 'Change %']]
y = data['Target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from sklearn.ensemble import RandomForestRegressor


model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python

predictions = model.predict(X_test)


from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f'MSE: {mse}')
print(f'R2 Score : {r2}')
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(16, 6))


plt.subplot(1, 2, 1)
plt.plot(y_test.index, y_test, color='blue', label='Real Prices')
plt.title('Real Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(y_test.index, predictions, color='red', label='Predicted Prices', linestyle='--')
plt.title('Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()


plt.tight_layout()
plt.show()

```
